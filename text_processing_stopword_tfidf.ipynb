##1(c) What are the limitations of this data processing flow? How can we overcome those?

Limitations of the current data processing flow
- Limitations of N-grams
1. Data Sparsity

Example: If n=3 (trigrams), many word combinations may appear very rarely, making the model unreliable.

Solution: Use backoff and smoothing techniques (like Laplace smoothing) to handle missing n-grams.

2. Out-of-Vocabulary (OOV) Words

Example: If a test dataset contains new words (e.g., "COVID-19" in 2020), an n-gram model trained before 2020 won’t recognize it.

Solution: Use subword tokenization (like Byte Pair Encoding - BPE) to break words into smaller meaningful parts.

3. Contextual Limitation

Example: "Apple" in "I ate an apple" vs. "Apple released a new iPhone" are treated as identical.

Solution: Use word embeddings (Word2Vec, GloVe, or BERT) instead of n-grams to capture context.

4. Long-Distance Dependencies

Example: In "The cat, which was very hungry, ate the fish," an n-gram model might fail to connect "cat" and "ate" if n is too small.

Solution: Use transformer-based models (BERT, GPT) which capture long-term dependencies.

- Limitations of TF-IDF
1. Inability to Capture Synonyms

Example: "happy" and "joyful" are treated as completely different words.

Solution: Use word embeddings (Word2Vec) to capture semantic similarity.

2. Ignoring Word Order

Example: "not good" and "good not" would have the same TF-IDF score, even though they mean different things.

Solution: Use n-grams in combination with TF-IDF to capture some order information.

3. Bias Towards Long Documents

Example: A long document will naturally have higher TF values, making it seem more important than a shorter but more relevant one.

Solution: Use normalization techniques like BM25 instead of plain TF-IDF.

## Term frequency - Invert document frequency (TF-IDF)


### TF
Term frequency (TF) is the number of times a given term appears in document

$$
tf(t) = f(t,d)\times\frac{1}{T}
$$
whereas, $f(t,d)$ is the frequency of the word $t$ in the document $d$, $T$ is the number of all words in that document.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]

# Declare TF vectorize
tf_vectorizer = TfidfVectorizer(ngram_range=(1, 1),
                                use_idf=False, # only using TF
                                norm='l1')

tf_vectorizer.fit_transform(corpus)

tf_vectorized = tf_vectorizer.transform(corpus)

tf_output = tf_vectorized[0]

# Build TF table
words_tf_idf = pd.DataFrame(tf_output.T.todense(), index=tf_vectorizer.get_feature_names_out(), columns=['tf'])
words_tf_idf
```

### IDF

Inverse Document Frequency, or abbreviated as IDF, measures how important a term is. While computing TF, all terms are considered equally important. However it is known that certain terms, such as "is", "of", and "that", may appear a lot of times but have little importance. Thus we need to weigh down the frequent terms while scale up the rare ones.

$$
idf(t) = \log\left(\frac{\text{#documents in the document set}}{\text{#documents with term}}\right) + 1
$$

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]

# Configure settings for IDF vectorize
tf_idf_vectorizer = TfidfVectorizer(ngram_range=(1, 1),
                                    smooth_idf=False,
                                    use_idf=True,
                                    norm=None)

tf_idf_vectorizer.fit_transform(corpus)

# Retrieve only idf information
idf_vectorizer = tf_idf_vectorizer.idf_

# Join idf values into the previous dataframe
words_tf_idf['idf'] = idf_vectorizer

# Show dataframe with ascending values of idf
words_tf_idf.sort_values(by=['idf'])
```

### TF-IDF

Technically saying, TF-IDF is a score which is applied to every word in every document in our dataset. And for every word, the TF-IDF value increases with every appearance of the word in a document, but is gradually decreased with every appearance in other documents

$$
\text{tf-idf}= tf(t, d) \times idf(t)
$$

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]

tf_idf_vectorizer = TfidfVectorizer(ngram_range=(1, 1),
                                    smooth_idf=False,
                                    use_idf=True,
                                    norm='l1')

tf_idf_vectorizer.fit_transform(corpus)

tf_idf_vectorized = tf_idf_vectorizer.transform(corpus)

tf_idf_output = tf_idf_vectorized[0]
words_tf_idf['tf-idf'] = tf_idf_output.T.todense()

words_tf_idf.sort_values(by=['tf-idf'])
```

##  (2a) Build the tf-idf table for the UIT-VSFC dataset with $n$-gram = 1 and $n$-gram = 2.

```python
def compute_tf_idf(corpus: list, n=1, smooth_idf=False, sublinear_tf=False, norm='l2') -> pd.DataFrame:
    """
    Compute the TF-IDF table for a given corpus.

    Args:
        corpus (list): List of sentences/documents.
        n (int): The n-gram value to consider (default is unigram).
        smooth_idf (bool): Whether to smooth IDF weights by adding 1 to document frequencies (default is False).
        norm (str): Normalization to apply to term vectors (default is None).
        sublinear_tf (bool): Whether to use sublinear term frequency scaling (default is False).

    Returns:
        pd.DataFrame: A DataFrame containing the TF-IDF matrix.
    """

    # Configure TfidfVectorizer with the given parameters
    vectorizer = TfidfVectorizer(
        ngram_range=(n, n),
        smooth_idf=smooth_idf,
        use_idf=True,
        norm=norm,
        sublinear_tf=sublinear_tf
    )

    # Fit and transform the corpus to get TF-IDF matrix directly
    tf_idf_matrix = vectorizer.fit_transform(corpus)

    # A
    idf_values = vectorizer.idf_

    # Convert the TF-IDF matrix to a dense array (if needed) and then to DataFrame
    tf_matrix = tf_idf_matrix.copy()  # Here tf_idf_matrix already has the effect of TF*IDF


    # Convert the matrix to a dense array  and then to DataFrame
    df = pd.DataFrame(tf_idf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
    return df
```

TF-IDF Hyperparameters
======================================

When processing text with TF-IDF (Term Frequency-Inverse Document Frequency), several hyperparameters can be adjusted. These include **smooth_idf**, **sublinear_tf**, and **norm**. The following explanation describes each parameter with simple language and examples suitable for a high school level, without using second-person pronouns.

Smooth_idf
----------
**Functionality:**  
This parameter adds a small constant to the count of documents in which a term appears. It prevents rare terms from obtaining an excessively high weight.

**Analogy:**  
Imagine grading homework submissions.  
- Without smoothing: A student who submits homework only once might receive an unusually high score for uniqueness.  
- With smoothing (`smooth_idf=True`): A small bonus is added to every student's score, ensuring that a one-time submission does not stand out excessively.

Sublinear_tf
------------
**Functionality:**  
Instead of using the raw frequency count of a term, this parameter applies a logarithmic scaling. This reduces the impact of very frequent terms.

**Analogy:**  
Consider counting candies in jars.  
- Without sublinear scaling: A jar containing 100 candies appears to be 10 times larger than a jar with 10 candies.  
- With sublinear scaling (`sublinear_tf=True`): The difference is less dramatic, balancing the influence of jars with many candies compared to those with fewer candies.

Norm
----
**Functionality:**  
Normalization scales the TF-IDF scores so that document vectors are comparable, regardless of the number of words or document length.

**Analogy:**  
Think of comparing test scores from two classes—one class with tests out of 100 points and another with tests out of 50 points.  
- **With normalization (e.g., 'l1' or 'l2'):** Scores are converted to a common scale, similar to expressing them as percentages.  
- **Without normalization (None):** Documents with more words or higher term frequencies may have inflated scores, which does not necessarily reflect better performance.

Summary
-------

- **Smooth IDF:**  
  - *Enabled:* Rare terms receive a more balanced weight.  
  - *Disabled:* Rare terms might obtain disproportionately high scores.
- **Sublinear TF:**  
  - *Enabled:* The influence of very frequent terms is moderated.  
  - *Disabled:* Frequent terms can dominate the scoring.
- **Normalization (Norm):**  
  - *Applied:* Document vectors are scaled for fair comparison.  
  - *Not applied:* Longer documents or those with repetitive content may exhibit inflated scores.

```python
def get_tf_idf_extremes(df):
    """
    Find the word with the lowest and highest TF-IDF values, and print the list of words with the minimum TF-IDF value.

    Args:
        df (pd.DataFrame): DataFrame containing TF-IDF values for words.

    Returns:
        tuple: A tuple containing:
            - max_word (str): Word with the highest TF-IDF value.
            - min_value (float): The lowest TF-IDF value.
            - max_value (float): The highest TF-IDF value.
            - min_words_list (list): List of words with the lowest TF-IDF value.
    """
    # Word with lowest TF-IDF
    min_value = df.min().min()
    min_words_list = df.columns[df.min() == min_value].tolist()

    # Word with highest TF-IDF
    max_word = df.max().idxmax()
    max_value = df.max().max()

    return max_word, min_value, max_value, min_words_list
```

At their core, TF-IDF and n-grams serve different purposes:
-----------------------------------
- TF-IDF is a weighting method applied to terms to assign importance based on frequency and rarity across documents.

- N-grams are a representation method for extracting sequences of words from text.
However, they are often used together: n-grams (unigrams, bigrams, etc.) can be the "terms" to which TF-IDF is applied. For example, you might calculate TF-IDF scores for bigrams like "love dogs" instead of just unigrams like "dogs". This suggests they are complementary rather than entirely separate.

# **2d**) Which limitations from $n$-grams that TF-IDF overcame ?


Sparsity:
-----------------------------------
How TF-IDF Helps: While TF-IDF doesn’t eliminate sparsity, it reduces its impact by assigning low weights to rare n-grams that appear in few documents (via low IDF scores). This shifts the focus to more informative n-grams, mitigating the problem of sparse, infrequently occurring sequences.

High Dimensionality:
------------------------------------
How TF-IDF Helps: TF-IDF itself doesn’t reduce the number of features, but it enables effective feature selection. By prioritizing n-grams with high TF-IDF scores, less informative ones can be filtered out, indirectly addressing the high-dimensionality issue when paired with such techniques.
Sensitivity to Noise:
-------------------------------------
How TF-IDF Helps: By focusing on term importance rather than just frequency, TF-IDF can be somewhat more robust to noise. For example, if "love dogs" and "adore dogs" both contain "dogs" as a key term, TF-IDF might still highlight "dogs" as important, even if the exact n-grams differ due to synonyms or variations.
Limitations TF-IDF Does Not Overcome
---------------------------
TF-IDF doesn’t fully resolve all n-gram limitations:

Lack of Semantic Understanding: TF-IDF still treats n-grams as discrete units and cannot recognize semantic similarities (e.g., "not good" vs. "bad").
Context Beyond N-grams: It doesn’t capture word order or long-range dependencies beyond the chosen n, as it relies on a bag-of-words-like approach.


Contextual understanding: While n-grams capture some context by considering word sequences, TF-IDF goes beyond simple co-occurrence.  By incorporating IDF, it downweights frequent words (like "the" or "a") that appear in many documents but carry less specific meaning.  This implicitly emphasizes terms that are more distinctive to a particular document or topic.

Dimensionality reduction:  N-grams, especially higher-order ones, lead to a very high-dimensional feature space.  TF-IDF itself doesn't reduce dimensionality, but it assigns lower weights to less important features, making the feature space more manageable when used with dimensionality reduction techniques (like PCA or SVD) for downstream applications.  It helps to focus on the most relevant features.

Handling of frequent but less informative words: N-grams treat all words equally based on their frequency. TF-IDF addresses the issue of frequent, uninformative words by down-weighting them with IDF.  N-grams might overrepresent these words, making it difficult to discern the true topic or meaning of the document.

In essence, TF-IDF enhances n-grams by adding a weighting scheme that distinguishes between important and unimportant words within a document relative to a collection of documents.  It uses the inverse document frequency component to improve upon the limitations of purely counting n-gram occurrences.
